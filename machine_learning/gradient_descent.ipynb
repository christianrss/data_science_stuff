```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5) ** 2 + 3
print(X,Y)
```

```python
# Visualize
#plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# Gradient Descent
# init x with any random value
x = 0
lr = 0.1 # learn rate
for i in range(50):
    grad = 2 * (x-5)
    x = x - lr * grad
    print(x)
```
