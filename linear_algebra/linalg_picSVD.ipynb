```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
pic = Image.open('images/kasparov.jpg')

plt.imshow(pic)
plt.show()

pic = pic.convert('L')

pic = np.array(pic)
plt.imshow(pic,cmap='gray')
plt.show()
```

```python
# SVD (singular value decomposition)
U,S,V = np.linalg.svd(pic)

# plot the spectrum
plt.plot(S, 's-')
plt.xlim([0,100])
plt.xlabel('Component number')
plt.ylabel('Singular value (\sigma)')
plt.show()
```

```python
# reconstruct the image based on some components

# list the components to use for the reconstruction
comps = np.arange(10,60)

#reconstruct the low-rank version of the picture
reconPic = U[:,comps]@np.diag(S[comps])@V[comps,:]

#show the original and reconstructed pictures for comparison
plt.figure(figsize=(5,10))
plt.subplot(1,2,1)
plt.imshow(pic,cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(reconPic,cmap='gray')
plt.title('Components %s-%s' %(comps[0],comps[-1]))
plt.axis('off')
plt.show()
```
